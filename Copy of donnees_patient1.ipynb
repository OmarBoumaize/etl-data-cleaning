```python
def fix_date(date):
  #first we need to make sure that the date is corrected format
  verify_date = re.search(r'(\d{1,2}/\d{1,2}/\d{3,4})' ,str(date))
  if verify_date:
    date = verify_date.group(1)
    extracted_year  = re.search(r'\d{1,2}/\d{1,2}/(\d{3,4})' , date)
    extracted_year = extracted_year.group(1)

    #veriy if the year have number have issues like thsi 119 -> 1919
    three_digits= re.search(r'^(\d{3})$' , extracted_year)
    if  three_digits  and (int(re.search(r'(\d)\d{2}' , extracted_year).group(1))) < 9  :
        extracted_year = re.sub(r'(\d)(\d)(\d)' , lambda x :  x.group(1)+"m" +  x.group(3)  , extracted_year )
        return re.sub(r'(\d{1,2}/\d{1,2}/)(\d{3})' , lambda x : x.group(1) + str(extracte_year) , date )
    # verify if the year have 9 in the beginning of  3 digits so we will just add 1
    if three_digits and re.search(r'(\d)\d{2}' , extracted_year).group(1) == "9" :
      extracted_year =  "1" + extracted_year
      return re.sub(r'(\d{1,2}/\d{1,2}/)(\d{3})' , lambda x : x.group(1) + str(extracte_year)  , date)




    # verify if the number now have 4 digits





    four_digits = re.search(r'^(\d{4})$' , extracted_year)


    if four_digits and (int(re.search(r'(\d)(\d)(\d{2})', extracted_year).group(2)) < 9 and int(re.search(r'(\d)(\d{3})', extracted_year).group(1)) != 1):
         # Replace first digit with '1' and second with '9' in one go
         extracted_year = re.sub(r'^(\d)(\d)(\d{2})$', lambda x: '1' + '9' + x.group(3), extracted_year)
         return re.sub(r'(\d{1,2}/\d{1,2}/)(\d{4})', lambda x: x.group(1) + extracted_year, date)

    if  four_digits and (   (int(re.search(r'(\d)(\d{3})' , extracted_year).group(1) )) !=1 ):
      extracted_year = re.sub(r'(\d)(\d{3})' ,lambda x : "1" + x.group(2) , extracted_year)
      return re.sub(r'(\d{1,2}/\d{1,2}/)(\d{4})' , lambda x : x.group(1) + str(extracted_year) , date )


    if   four_digits and  (int(re.search(r'(\d)(\d)(\d{2})' , extracted_year).group(2) )) < 9 :
       extracted_year = re.sub(r'(\d)(\d)(\d{2})' ,lambda x : x.group(1) + "9"  + x.group(3) , extracted_year)
       return re.sub(r'(\d{1,2}/\d{1,2}/)(\d{4})' , lambda x : x.group(1) + str(extracted_year) , date )



    return date
fix_date('hdh	01/01/1955hdhshs7')

test = data_patient1[['dnaiss' , 'year_naiss']][data_patient1['year_naiss'] < 1900 ]

test['after_fix_date'] = test['dnaiss'].apply(fix_date)

#fix_date('hdh	01/11/0195hdhshs7')


test
```
