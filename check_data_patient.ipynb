```python
def search_cin(text):
    if isinstance(text, str):
        text = text.replace(' ' , '').upper()
        pattern_with_cin = r'CIN.*([A-Za-z]{1,2}\d{6,7})'
        match = re.search(pattern_with_cin, text, re.IGNORECASE)
        if match:
            return True
        else :
          pattern_without_cin = r'([A-Za-z]{1,2}\d{5,7})'
          match = re.search(pattern_without_cin  , text)
          if match :
            return True
    return False


def extract_cin(text):
    text = str(text).upper().replace(' ', '')

    # Patterns that require "CIN" prefix followed by valid code (letters+digits)
    patterns_with_cin = [
        r'CIN[:.\s]*([A-Z]{2}\d{6})',
        r'CIN[:.\s]*([A-Z]{2}\d{5})',
        r'CIN[:.\s]*([A-Z]{1}\d{7})',
        r'CIN[:.\s]*([A-Z]{1}\d{6})',
        r'CIN[:.\s]*([A-Z]{1}\d{5})'
    ]

    # Patterns without CIN prefix
    patterns_without_cin = [
        r'([A-Z]{2}\d{6})',
        r'([A-Z]{2}\d{5})',
        r'([A-Z]{1}\d{7})',
        r'([A-Z]{1}\d{6})',
        r'([A-Z]{1}\d{5})'
    ]

    #  match with CIN prefix first
    for pattern in patterns_with_cin:
        match = re.search(pattern, text)
        if match:
            return match.group(1)

    # Then  match without CIN prefix
    for pattern in patterns_without_cin:
        match = re.search(pattern, text)
        if match:
            return match.group(1)

    return np.nan

# Testing
print(extract_cin('68LOT AZELF ESSAOUIRA CIN:N13458'))



```

```python
def search_fax(text):
  if isinstance(text , str):
    text = text.replace(' ' , '')
    patterns_phone =r'\D'
    filtered = re.sub(patterns_phone,'' , text)
    extract_fax = re.search(r'(\d{9})' , filtered)
    if extract_fax :
      return True
  return False


def extract_fax(text):
  if isinstance(text , str):
    text = text.replace(' ' , '')
    patterns_phone =r'\D'
    filtered = re.sub(patterns_phone,'' , text)
    pattern_fax1 = re.search(r'(?:212)*[+*.:/,-]*(0\d{9})$' , filtered)
    pattern_fax2 = re.search(r'(?:212)*[+*.:/,-]*(\d{9})$' , filtered)

    if pattern_fax1 :
      return pattern_fax1.group(1)

    elif pattern_fax2 :
      return '0' + pattern_fax2.group(1)

  return np.nan
print(extract_fax('CIN : N 159477 / 670249049'))
print(extract_fax('CIN : N 159477 / 212+0670249049'))
```
