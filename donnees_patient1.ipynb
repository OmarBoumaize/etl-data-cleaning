```python
def extract_first_name(text):


  if isinstance(text , str):

      names_morrocain_pattern = r'(^(([A-Za-z]+\s+)?\s*(el|ait|ait sidi|sidi)\s+[A-Z]+))'


      match = re.search(names_morrocain_pattern , text , re.IGNORECASE)
      text_split = text.split()
      if len(text_split)== 2 :
        return (text_split[0])
      elif match :
        return match.group(1)

      elif len(text_split)== 2 :
        return (text_split[0])

      elif len(text_split) ==4 or len(text_split) ==5 :
        return " ".join(text_split[0:len(text_split) -1])

      return text_split[0]

  return text




def extract_last_name(text):


  if isinstance(text , str) :
    names_morrocain_pattern = r'((?:el|ait|sidi|moulay)\s+[A-Za-z]+|[A-Za-z]+)$'


    match = re.search(names_morrocain_pattern , text , re.IGNORECASE)
    text_split = text.split()
    if len(text_split)== 2 :
      return (text_split[1] )
    elif match :
      return match.group(1)

    return text_split[-1]

  return text


print(extract_last_name('EL MASLOHI EL BATOUL') )
print(extract_last_name('EL OUAD MOHAMMED MOUHCINE') )


data_patient['nom'] = data_patient['nomprenom'].apply(extract_first_name)
data_patient['prenom'] = data_patient['nomprenom'].apply(extract_last_name)
#data_patient[['nom' , 'prenom', 'nomprenom']][data_patient['prenom']==False]



```

```python
import pandas as pd
import joblib
import re
from sklearn.metrics import accuracy_score
import numpy as np # Import numpy

# Re-load the model and define helper functions (assuming they are correct)
model = joblib.load('/content/model.joblib')

def data_frame_user(list_names):

    def get_custom_features(list_name):
        ends_with_a = []
        ends_with_ia = []

        for name in list_name:

            if pd.notnull(name) and isinstance(name, str) :
                ends_with_a.append(1 if name.endswith('A') else 0)
                ends_with_ia.append(1 if name.endswith('IA') else 0)
            else:
                # If not a string or NaN, append 0 for both features
                ends_with_a.append(0)
                ends_with_ia.append(0)

        return ends_with_a, ends_with_ia


    new_name_features = get_custom_features(list_names)


    data = []
    for i in range(len(list_names)):
        data.append({
            "Name": list_names[i],
            "ends_with_a": new_name_features[0][i],
            "ends_with_ia": new_name_features[1][i]
        })

    return pd.DataFrame(data)


def gender_prediction_based_on_name(list_names):
  # Ensure list_names does not contain NaN values
  # Replace NaN with empty strings or drop them based on your needs
  cleaned_list_names = ["" if pd.isna(name) else name for name in list_names]

  data = data_frame_user(cleaned_list_names) # Use the cleaned list
  proba = model.predict_proba(data)
  proba_female =  model.predict_proba(data)[:,0]
  df_proba_female = pd.DataFrame(proba_female , columns = ["proba_female"] , index=data.index)
  results = pd.concat([data["Name"] , df_proba_female] , axis = 1)
  results["predicted_gender"] = results["proba_female"].apply(lambda x : "M" if x < 0.63 else "F")
  return results

# Re-create patient_names and clean it
patient_names = data_patient['prenom'].tolist()

# Now, call the function with the potentially large list
data_patient['predicted_gender'] = gender_prediction_based_on_name(patient_names)['predicted_gender']
```
